```python
import numpy as np
import xarray as xr
import dask
from graphviper.graph_tools.coordinate_utils import make_parallel_coord, make_frequency_coord
from graphviper.graph_tools import map, reduce
from astroviper._domain._image._moment_max import _moment_max
from astroviper.imaging._utils import _make_image
from graphviper.utils.display import dict_to_html
from IPython.display import HTML, display

parallel_coords = {}

###############################
coord = ps['uid___A002_Xf07bba_Xbe5c_target.lsrk_ddi_27_intent_OBSERVE_TARGET#ON_SOURCE_field_id_3'].time
parallel_coords["time"] = make_parallel_coord(coord=coord, n_chunks=24)

# ps_freq_axis = ps.get_ps_freq_axis()
# parallel_coords["frequency"] = make_parallel_coord(coord=ps_freq_axis, n_chunks=24*4)
###############################

# coord = ps['uid___A002_Xf07bba_Xbe5c_target.lsrk_ddi_27_intent_OBSERVE_TARGET#ON_SOURCE_field_id_3'].time.isel(time=slice(0,10))
# parallel_coords["time"] = make_parallel_coord(coord=coord, n_chunks=1)

from graphviper.graph_tools.coordinate_utils import interpolate_data_coords_onto_parallel_coords
node_task_data_mapping = interpolate_data_coords_onto_parallel_coords(parallel_coords, ps)
display(HTML(dict_to_html(node_task_data_mapping)))

# for xds_name in node_task_data_mapping[0]['data_selection']:
#     node_task_data_mapping[0]['data_selection'][xds_name]['frequency'] = slice(0,10) 
# node_task_data_mapping[0]['parallel_dims'].append('frequency')
# display(HTML(dict_to_html(node_task_data_mapping)))

def calc_stats(input_params):
    import graphviper.utils.logger as logger
    from xradio.vis.load_processing_set import load_processing_set, processing_set_iterator
    ps_iter = processing_set_iterator(input_params["data_selection"], input_params["input_data_store"], input_params["input_data"], data_variables=['VISIBILITY'], load_sub_datasets=False)
    
    min_vals = []
    max_vals = []
    for ms_xds in ps_iter:
        vis_data = ms_xds.VISIBILITY.values.ravel()
        max_val = vis_data[np.nanargmax(np.abs(vis_data))]
        min_val = vis_data[np.nanargmin(np.abs(vis_data))]
        min_vals.append(min_val)
        max_vals.append(max_val)

    result = np.array([np.max(max_vals),np.min(min_vals)])
    logger.info('The max is for this chunk: ', result)
    return result


input_params = {}
input_params['input_data_store'] = ps_name

import time
viper_graph = map(
    input_data=ps,
    node_task_data_mapping=node_task_data_mapping,
    node_task=calc_stats,
    input_params=input_params,
    in_memory_compute=False
)

def stats_reduce(graph_inputs, input_params):
    import numpy as np
    stats_array = np.array(graph_inputs) #graph_inputs is a list of 2-element arrays (max and min values).
    max_val = stats_array[np.nanargmax(np.abs(stats_array[:,0])),0]
    min_val = stats_array[np.nanargmin(np.abs(stats_array[:,1])),1]
    return np.array([max_val,min_val])

input_params = {}
viper_graph = reduce(viper_graph, stats_reduce, input_params, mode='tree') # mode "tree","single_node

from graphviper.graph_tools import generate_dask_workflow
dask_graph = generate_dask_workflow(viper_graph)

```

```python
start = time.time()
max_min=dask.compute(dask_graph)
print(time.time()-start)
max_min
```
